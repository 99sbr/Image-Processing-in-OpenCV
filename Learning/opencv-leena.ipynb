```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
% matplotlib inline
```

```python
img = cv2.imread('Lenna.png')
img1 = cv2.imread('Lenna.png')
plt.imshow(img)
plt.plot()
plt.show()
```

```python
img2=cv2.rectangle(img1,(250,400),(350,300),(0,0,255),10)
plt.imshow(img2)
plt.show()
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
img3=cv2.putText(img2,'Heyy Lenna!',(0,130), font, 1, (100,120,255), 2, cv2.LINE_AA)
plt.imshow(img3)
```

```python
img4=cv2.add(img3,img3)
plt.imshow(img4)
```

```python
face=img[200:400,200:350]
plt.imshow(face)
```

```python
img4[200:400,200:350]=face
plt.imshow(img4)
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img4,-1,kernel)

plt.subplot(121),plt.imshow(img4),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)
plt.imshow(blur)
```

```python
median = cv2.medianBlur(img4,5)
plt.subplot(121),plt.imshow(img4),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('median')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```
